```r

# Remove the outliers form the data set for linear regression model
outliers <- which(combined_df$longitude < -114.001)
length(outliers)
new_combined <- combined_df[-outliers,]
rownames(new_combined) = 1:5772
dim(new_combined)

# Linear regression model
lm_new_combined <- lm(new_combined$latitude~new_combined$longitude)
# Plot the regression line
plot(new_combined$longitude,new_combined$latitude)
abline(lm_new_combined)
```

```r
# Please install sp and rgdal packages
library(sp)
library(rgdal)
# This following step is similar to transforming time stamps like "2020-01-25" into a time object
spat_df_all <- SpatialPointsDataFrame(coords=new_combined[, c("longitude", "latitude")],
                                  data=new_combined['time'],   # This needs to be a data frame
                                  proj4string=CRS("+proj=longlat +datum=WGS84"))
# This step converts the longitude/latitude -> UTM
utm_df_all <- spTransform(spat_df_all, CRSobj = "+proj=utm +zone=12 +datum=WGS84")
utm_coords_all <- coordinates(utm_df_all)
```
